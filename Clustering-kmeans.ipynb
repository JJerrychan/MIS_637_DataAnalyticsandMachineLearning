### Load Dataset

```python
import pandas as pd
df = pd.read_csv('Live.csv')
print('Dataset dims:',df.shape)
print(df.info())
df.head()
```

### Remove Columns with Missing Data
Columns 1,2,3 and 4 have 100% missing data. Note that they are not listed in the data attributes coming from the data source.  
We need to drop these columns.

```python
df = df.drop(['Column1','Column2','Column3','Column4'],axis = 1)
print('Dataset dims:',df.shape)
print(df.info())
df.head()
```

### Which Categorical Variables Are Useful?
status_id, status_type, status_published are the three categorical fields in this dataset.  
Status_id seems like a id number for each post and status_published seems like the time of publication. So, these two are not distinguishing features of the posts. They basically indentify most posts uniquly.  
On the other hand, status_type is the type of a post. It could be useful in grouping posts together. 

we can used a counter to encode status_type values.

```python
from collections import Counter
print('Count of unique values in status_id column:',len(df['status_id'].unique()))
type_counter = Counter(df['status_type'])
print('Count of unique values in status_published column:',
      len(df['status_published'].unique()))
print('status_type',type_counter)
```

Let's drop status_id and status_published and encode status_type.

```python
df = df.drop(['status_id','status_published'],axis = 1)

# Do you remeber encoding using Counter
df['status_type'] = df['status_type'].apply(lambda x: list(type_counter.keys()).index(x))

print(df.info())
df.head()
```

### Normalization: MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[df.columns] = scaler.fit_transform(df[df.columns])

print(df.shape)
print(df.info())
df.head()
```

### k-means: k = 2

```python
from sklearn.cluster import KMeans

k2means = KMeans(n_clusters=2, random_state=42) 

k2means.fit(df)
```

### Centroids

```python
k2means.cluster_centers_
```

### Within-Cluster Variation

```python
k2means.inertia_
```

### Elbow Method for Determining k

```python
import matplotlib.pyplot as plt
SSEs = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 42)
    kmeans.fit(df)
    SSEs.append(kmeans.inertia_)
plt.plot(range(1, 11), SSEs)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('CS')
plt.show()
```
